# Exercício Proposto

Continue a explorar e faça gráficos pertinentes para a análise das notas dos alunos, com variáveis criadas no exercício da aula anterior

```python
n = 200

# Gerando nomes aleatórios masculinos e femininos (50% cada um)
nomesM = [names.get_first_name(gender = 'male') for i in range(int(n/2))]
nomesF = [names.get_first_name(gender = 'female') for i in range(int(n/2))]

nomes = nomesM + nomesF

# Gerando uma lista do sexo masculino e feminino cada uma igual a n/2
sexo = ['M']*int(n/2) + ['F']*int(n/2) 

# Gerando notas aleatórias para três provas
p1 = np.random.uniform(0, 7, n)
p2 = np.random.uniform(3, 10, n)
p3 = np.random.uniform(1, 9, n)

# Gerando Dados
dicionario = {'Nomes': nomes,
              'Sexo': sexo,
              'P1': p1,
              'P2': p2,
              'P3': p3}

notas = pd.DataFrame(dicionario).round(1)

# Gerando coluna de Média entre as Provas
notas['MF'] = np.ceil(10*notas.mean(axis=1))/10

# Gerando coluna de frequência
notas['Frequência'] = np.random.uniform(0, 100, n).round(1)

# Gerando Coluna Aprovado/Reprovado por Nota

def auxiliar(nota):
  if nota >= 6:
    saida = 'Aprovado'
  else:
    saida = 'RN'
  return saida

notas['Situação'] = notas['MF'].apply(auxiliar)

# Gerando Coluna Reprovados por Falta

def auxiliar2(RepFreq):
  if RepFreq < 75:
    saida2 = 'RF'
  else:
    saida2 = 'Não'
  return saida2

notas['Reprovação por Falta?'] = notas['Frequência'].apply(auxiliar2)

# Gerando grupos de frequências

def auxiliar3(frequencia):
  if frequencia <= 40:
    saida3 = 'Baixa Frequência'
  if frequencia > 40 and frequencia < 70:
    saida3 = 'Média Frequência'
  if frequencia >= 70:
    saida3 = 'Alta Frequência'
  return saida3

notas['Situação Frequência'] = notas['Frequência'].apply(auxiliar3)
```

```python
RN_M = notas[notas['Situação'] == 'RN'][notas['Sexo'] == 'M']['Nomes'].count()
AP_M = notas[notas['Situação'] == 'Aprovado'][notas['Sexo'] == 'M']['Nomes'].count()
RN_F = notas[notas['Situação'] == 'RN'][notas['Sexo'] == 'F']['Nomes'].count()
AP_F = notas[notas['Situação'] == 'Aprovado'][notas['Sexo'] == 'F']['Nomes'].count()

RN = [RN_M, RN_F]
AP = [AP_M, AP_F]

index = ['M', 'F']

df = pd.DataFrame({'Reprovados': RN,
                   'Aprovados': AP}, index=index)


ax = df.plot.barh(color={"Reprovados": "red", "Aprovados": "green"})
```
