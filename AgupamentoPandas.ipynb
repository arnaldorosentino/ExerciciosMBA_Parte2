# Exercício Proposto

Crie uma coluna no conjunto de dados que represente a presença do aluno durante o curso. Pode ser uma variável gerada através de uma distribuição uniforme, que represente a frequência em porcentagem do aluno.

Diante disso:

- Calcule a média final dos grupos: pouca frequencia (até 40%), média frequencia (entre 40% e 70%) e muito frequente (acima de 70%)
- Um aluno reprova por nota se sua frequencia é inferior a 75%. Verifique quantos alunos foram reprovados apenas pelo fato de não ter frequência suficiente.
- Qual a nota final média, por sexo, de quem não reprovou por falta, mas reprovou por nota? E de quem reprovou por falta, mas não reprovou por nota?

```python
n = 200

nomesM = [names.get_first_name(gender = 'male') for i in range(int(n/2))]
nomesF = [names.get_first_name(gender = 'female') for i in range(int(n/2))]
```

```python
nomes = nomesM + nomesF

# Gerando uma lista do sexo masculino e feminino cada uma igual a n/2
sexo = ['M']*int(n/2) + ['F']*int(n/2) 

p1 = np.random.uniform(0, 10, n)
p2 = np.random.uniform(3, 10, n)
p3 = np.random.uniform(1, 9, n)

dicionario = {'Nomes': nomes,
              'Sexo': sexo,
              'P1': p1,
              'P2': p2,
              'P3': p3}

notas = pd.DataFrame(dicionario).round(1)


notas['MF'] = np.ceil(10*notas.mean(axis=1))/10

# Gerando coluna de frequência
notas['Frequência'] = np.random.uniform(0, 100, n).round(1)

notas
```

```python
# Gerando Coluna Aprovado/Reprovado por Nota

def auxiliar(nota):
  if nota >= 6:
    saida = 'Aprovado'
  else:
    saida = 'RN'
  return saida

notas['Situação'] = notas['MF'].apply(auxiliar)

# Gerando Coluna Reprovados por Falta

def auxiliar2(RepFreq):
  if RepFreq < 75:
    saida2 = 'RF'
  else:
    saida2 = 'Não'
  return saida2

notas['Reprovação por Falta?'] = notas['Frequência'].apply(auxiliar2)

# Gerando grupos de frequências

def auxiliar3(frequencia):
  if frequencia <= 40:
    saida3 = 'Baixa Frequência'
  if frequencia > 40 and frequencia < 70:
    saida3 = 'Média Frequência'
  if frequencia >= 70:
    saida3 = 'Alta Frequência'
  return saida3

notas['Situação Frequência'] = notas['Frequência'].apply(auxiliar3)
```

```python
# Atividade 1: Calcule a média final dos grupos: pouca frequencia (até 40%), média frequencia (entre 40% e 70%) e muito frequente (acima de 70%)

notas.groupby(['Situação Frequência']).mean()[['MF']]
```

```python
# Atividade 2: Um aluno reprova por nota se sua frequencia é inferior a 75%. 

# Verifique quantos alunos foram reprovados apenas pelo fato de não ter frequência suficiente.

# Opção 1
notas.groupby('Reprovação por Falta?').describe()['MF']

# Opção 2
notas['MF'].groupby(notas['Reprovação por Falta?']).count()

```

```python
# Atividade 3: Qual a nota final média, por sexo, de quem não reprovou por falta, mas reprovou por nota? 
#              E de quem reprovou por falta, mas não reprovou por nota?

notas.groupby(['Situação', 'Reprovação por Falta?']).mean()[['MF']]
```
